# Differential Equations - Academic Notes

## Introduction to Differential Equations

A **differential equation** is a mathematical equation that relates a function $ y(x) $ to its derivatives. These equations are used to model real-world systems where change is involved, such as physics, biology, and engineering.

### General Form:
$
F\left(x, y, y', y'', \dots, y^{(n)}\right) = 0
$
where:
- $ y^{(n)} $: The $ n $-th derivative of $ y $.

## Types of Differential Equations

### 1. **Ordinary Differential Equations (ODEs):**
Equations involving a single independent variable.
$
y' + p(x)y = q(x)
$

### 2. **Partial Differential Equations (PDEs):**
Equations involving partial derivatives with multiple independent variables.
$
\frac{\partial^2 u}{\partial t^2} = c^2 \frac{\partial^2 u}{\partial x^2}
$
## Solving First-Order Differential Equations

### Example 1: $ y'(x) = y(x) $
Using separation of variables:
$
\frac{dy}{y} = dx
$
Integrating both sides:
$
\ln|y| = x + C \implies y(x) = Ce^x
$

### Example 2: $ y'(x) = \frac{1}{2y(x)} $
Rewriting:
$
2y\,dy = dx
$
Integrating:
$
y^2 = x + C \implies y(x) = \pm\sqrt{x + C}
$

## Separation of Variables for $ \frac{dy}{dx} = \frac{x}{y} $
$
y\,dy = x\,dx
$
Integrating both sides:
$
\frac{y^2}{2} = \frac{x^2}{2} + C \implies y^2 = x^2 + C
$

## Second-Order Differential Equations

### Example 1: $ y''(x) + y'(x) = 0 $
Characteristic equation:
$
m^2 + m = 0 \implies m(m + 1) = 0
$
Roots: $m = 0, -1 $
General solution:
$
y(x) = C_1 + C_2e^{-x}
$

### Example 2: $ y''(x) - y(x) = 0 $
Characteristic equation:
$
m^2 - 1 = 0 \implies m = \pm 1
$
General solution:
$
y(x) = C_1e^x + C_2e^{-x}
$

### Example 3: $ \frac{d^2y}{dx^2} = -\omega^2y(x) $
Harmonic oscillator solution:
$
y(x) = A\cos(\omega x) + B\sin(\omega x)
$

## Boundary Value Problems
For $ y''(x) + y'(x) = 0 $ with $ y(0) = 2 $, $ y'(0) = -1 $:
$
y(x) = 2 - x
$

## Partial Differential Equations: The Wave Equation
The general form:
$
\frac{\partial^2 \psi}{\partial t^2} = \nu^2 \frac{\partial^2 \psi}{\partial x^2}
$
Solution for $ \psi(t, x) = A\cos(\omega t + kx) $:
$
\frac{\partial^2 \psi}{\partial t^2} - \nu^2 \frac{\partial^2 \psi}{\partial x^2} = 0
$

Where $ \nu = \frac{\omega}{k} $.

## Conclusion
Differential equations are powerful tools in modeling and solving dynamic systems. By understanding the classification and methods of solution, one can address a wide range of real-world problems.

# Problem 1: Solve the first-order ordinary differential equations

---

## **1.1 Solve $y'(x) = y(x)$**

### Step 1: Identify the equation type
The given equation is:
$$
y'(x) = y(x).
$$
This is a **first-order linear ODE**.

### Step 2: Rewrite the equation
$$
\frac{dy}{dx} = y.
$$

### Step 3: Separate variables
Divide both sides by $y$:
$$
\frac{1}{y} \, dy = dx.
$$

### Step 4: Integrate both sides
$$
\int \frac{1}{y} \, dy = \int 1 \, dx.
$$
Results:
$$
\ln|y| = x + C.
$$

### Step 5: Solve for $y$
Exponentiate both sides:
$$
y = e^{x + C} = Ce^x, \quad \text{where } C = e^C.
$$

**Solution:**
$$
y(x) = Ce^x
$$

---

## **1.2 Solve $y'(x) = \frac{1}{2y(x)}$**

### Step 1: Rewrite the equation
$$
\frac{dy}{dx} = \frac{1}{2y}.
$$
Rearrange:
$$
2y \, dy = dx.
$$

### Step 2: Integrate both sides
$$
\int 2y \, dy = \int 1 \, dx.
$$
Results:
$$
y^2 = x + C.
$$

### Step 3: Solve for $y$
Take the square root:
$$
y = \pm\sqrt{x + C}.
$$

**Solution:**
$$
y(x) = \pm\sqrt{x + C}.
$$
$$
Left = (\pm\sqrt{x + C})'=\pm\frac{1}{2\sqrt{x + C}}
$$

$$
Right = \frac{1}{2y(x)} = \pm\frac{1}{2\sqrt{x + C}}
$$

# Problem 2: Solve the first-order ODEs using separation of variables

---

## **2.1 Solve $\frac{dy}{dx} = \frac{x}{y}$**

### Step 1: Rewrite the equation
$$
y \, dy = x \, dx.
$$

### Step 2: Integrate both sides
$$
\int y \, dy = \int x \, dx.
$$
Results:
$$
\frac{y^2}{2} = \frac{x^2}{2} + C.
$$

### Step 3: Simplify
Multiply through by 2:
$$
y^2 = x^2 + C.
$$

### Step 4: Solve for $y$
$$
y = \pm\sqrt{x^2 + C}.
$$

**Solution:**
$$
y(x) = \pm\sqrt{x^2 + C}.
$$

---

## **2.2 Solve $\frac{dy}{dx} = \frac{y}{x}$**

### Step 1: Rewrite the equation
$$
\frac{1}{y} \, dy = \frac{1}{x} \, dx.
$$

### Step 2: Integrate both sides
$$
\int \frac{1}{y} \, dy = \int \frac{1}{x} \, dx.
$$
Results:
$$
\ln|y| = \ln|x| + C.
$$

### Step 3: Solve for $y$
Exponentiate:
$$
|y| = e^{\ln|x| + C} = |x|e^C.
$$
Let $C_1 = e^C$:
$$
y = C_1x.
$$

**Solution:**
$$
y(x) = Cx.
$$

---

## **2.3 Solve $\frac{dy}{dx} = xy$**

### Step 1: Rewrite the equation
$$
\frac{1}{y} \, dy = x \, dx.
$$

### Step 2: Integrate both sides
$$
\int \frac{1}{y} \, dy = \int x \, dx.
$$
Results:
$$
\ln|y| = \frac{x^2}{2} + C.
$$

### Step 3: Solve for $y$
Exponentiate:
$$
y = e^{C}e^{x^2/2}.
$$
Let $C_1 = e^C$:
$$
y = C_1e^{x^2/2}.
$$

**Solution:**
$$
y(x) = Ce^{x^2/2}.
$$

# Problem 3: Solve the second-order ODEs

---

## **3.1 Solve $y''(x) + y'(x) = 0$, with $y(0) = 2$ and $y'(0) = -1$**

### Step 1: Characteristic equation
Assume $y(x) = e^{rx}$. Substituting gives:
$$
r^2 + r = 0.
$$
Factoring:
$$
r(r + 1) = 0 \implies r = 0, -1.
$$

### Step 2: General solution
$$
y(x) = C_1 + C_2e^{-x}.
$$

### Step 3: Apply boundary conditions
- At $y(0) = 2$:
$$
C_1 + C_2 = 2.
$$
- At $y'(0) = -1$:
$$
-C_2 = -1 \implies C_2 = 1.
$$

Substitute $C_2 = 1$:
$$
C_1 + 1 = 2 \implies C_1 = 1.
$$

**Solution:**
$$
y(x) = 1 + e^{-x}.
$$

---

## **3.2 Solve $y''(x) - y(x) = 0$, with $y(0) = 2$ and $y'(0) = 0$**

### Step 1: Characteristic equation
$$
r^2 - 1 = 0 \implies r = \pm 1.
$$

### Step 2: General solution
$$
y(x) = C_1e^x + C_2e^{-x}.
$$

### Step 3: Apply boundary conditions
- At $y(0) = 2$:
$$
C_1 + C_2 = 2.
$$
- At $y'(0) = 0$:
$$
C_1 - C_2 = 0 \implies C_1 = C_2.
$$

Substitute $C_1 = C_2 = 1$:
$$
y(x) = e^x + e^{-x}.
$$

**Solution:**
$$
y(x) = e^x + e^{-x}.
$$

---

## **3.3 Solve $y''(x) = -\omega^2y(x)$**

### Step 1: Characteristic equation
$$
r^2 + \omega^2 = 0 \implies r = \pm i\omega.
$$

### Step 2: General solution
$$
y(x) = C_1\cos(\omega x) + C_2\sin(\omega x).
$$

**Solution:**
$$
y(x) = C_1\cos(\omega x) + C_2\sin(\omega x).
$$

# Problem 4: Verify Wave Equation

---

### Wave Equation:
$$
\frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0.
$$
Given:
$$
\psi(t, x) = A\cos(\omega t + kx).
$$

### Step 1: Compute time derivatives
$$
\frac{\partial \psi}{\partial t} = -A\omega\sin(\omega t + kx),
$$
$$
\frac{\partial^2 \psi}{\partial t^2} = -A\omega^2\cos(\omega t + kx).
$$

### Step 2: Compute spatial derivatives
$$
\frac{\partial \psi}{\partial x} = -Ak\sin(\omega t + kx),
$$
$$
\frac{\partial^2 \psi}{\partial x^2} = -Ak^2\cos(\omega t + kx).
$$

### Step 3: Substitute into wave equation
$$
-A\omega^2\cos(\omega t + kx) - v^2(-Ak^2\cos(\omega t + kx)) = 0.
$$
Simplify:
$$
-A\omega^2 + Av^2k^2 = 0 \implies v^2 = \frac{\omega^2}{k^2}.
$$

**Verification Complete:** The function satisfies the wave equation.
