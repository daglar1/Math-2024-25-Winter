# Canonical Equations of Conic Sections

### 1. **Circle**

The canonical equation of a **circle** with center \( (h, k) \) and radius \( r \) is:

$$
(x - h)^2 + (y - k)^2 = r^2
$$

- **Example**: \( (x - 2)^2 + (y + 3)^2 = 9 \)

### 2. **Ellipse**

The canonical equation of an **ellipse** with center \( (h, k) \), semi-major axis \( a \), and semi-minor axis \( b \) is:

$$
\frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1
$$

- **Example**: \( \frac{(x - 1)^2}{4} + \frac{(y - 2)^2}{9} = 1 \)

### 3. **Hyperbola**

The canonical equation of a **hyperbola** with center \( (h, k) \), transverse axis length \( 2a \), and conjugate axis length \( 2b \) is:

$$
\frac{(x - h)^2}{a^2} - \frac{(y - k)^2}{b^2} = 1
$$

- **Example**: \( \frac{(x - 1)^2}{4} - \frac{(y - 2)^2}{9} = 1 \)

### 4. **Parabola**

The canonical equation of a **parabola** with vertex \( (h, k) \) and focal distance \( p \) is:

- Vertical parabola (opens up or down):

$$
(y - k) = a(x - h)^2
$$

- Horizontal parabola (opens left or right):

$$
(x - h) = a(y - k)^2
$$

- **Example**: \( y = x^2 \)

### 5. **General Form of a Conic Section**

A general conic section equation is:

$$
Ax^2 + Bxy + Cy^2 + Dx + Ey + F = 0
$$

- **Circle**: \( A = C \), \( B = 0 \)
- **Ellipse**: \( A \neq C \), \( B = 0 \)
- **Hyperbola**: \( B^2 - 4AC > 0 \)
- **Parabola**: \( B^2 - 4AC = 0 \)

### 6. **Foci and Eccentricity**

- **Circle**: \( e = 0 \)
- **Ellipse**: \( e = \sqrt{1 - \frac{b^2}{a^2}} \)
- **Hyperbola**: \( e = \sqrt{1 + \frac{b^2}{a^2}} \)
- **Parabola**: \( e = 1 \)

# Equations of Second-Order Curves with a Shifted Center

### 1. **Circle**

The canonical equation of a **circle** with a center at \( (h, k) \) and radius \( r \) is:

$$
(x - h)^2 + (y - k)^2 = r^2
$$

- **Example**: \( (x - 3)^2 + (y + 4)^2 = 16 \)

### 2. **Ellipse**

The canonical equation of an **ellipse** with center \( (h, k) \), semi-major axis \( a \), and semi-minor axis \( b \) is:

$$
\frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1
$$

- **Example**: \( \frac{(x - 2)^2}{9} + \frac{(y - 1)^2}{16} = 1 \)

### 3. **Hyperbola**

The canonical equation of a **hyperbola** with center \( (h, k) \), transverse axis length \( 2a \), and conjugate axis length \( 2b \) is:

$$
\frac{(x - h)^2}{a^2} - \frac{(y - k)^2}{b^2} = 1
$$

- **Example**: \( \frac{(x - 4)^2}{25} - \frac{(y - 5)^2}{16} = 1 \)

### 4. **Parabola**

The canonical equation of a **parabola** with vertex \( (h, k) \) and focal distance \( p \) is:

- Vertical parabola (opens up or down):

$$
y - k = a(x - h)^2
$$

- Horizontal parabola (opens left or right):

$$
x - h = a(y - k)^2
$$

- **Example**: \( y - 2 = 3(x - 1)^2 \)

### 5. **General Form of a Conic Section**

A general second-order conic section equation with a shifted center \( (h, k) \) is:

$$
A(x - h)^2 + B(x - h)(y - k) + C(y - k)^2 + D(x - h) + E(y - k) + F = 0
$$

- This equation represents a circle, ellipse, hyperbola, or parabola depending on the values of \( A \), \( B \), and \( C \).

### 6. **Foci and Eccentricity for Shifted Conics**

- **Circle**: \( e = 0 \)
- **Ellipse**: \( e = \sqrt{1 - \frac{b^2}{a^2}} \)
- **Hyperbola**: \( e = \sqrt{1 + \frac{b^2}{a^2}} \)
- **Parabola**: \( e = 1 \)

# Reduction of the General Second-Order Curve Equation to Canonical Form

### 1. **General Form of Second-Order Curve**

The general second-order equation of a conic section is:

$$
Ax^2 + Bxy + Cy^2 + Dx + Ey + F = 0
$$

Where:
- \( A, B, C, D, E, F \) are constants.
- The type of curve (circle, ellipse, hyperbola, or parabola) depends on the values of \( A \), \( B \), and \( C \).

### 2. **Steps to Reduce to Canonical Form**

#### **Step 1: Eliminate the \( Bxy \) Term (Rotation)**

To eliminate the \( Bxy \) term, we rotate the coordinate axes by an angle \( \theta \). The transformation equations are:

$$
x' = x \cos \theta + y \sin \theta
$$
$$
y' = -x \sin \theta + y \cos \theta
$$

The angle \( \theta \) is chosen such that:

$$
\tan(2\theta) = \frac{B}{A - C}
$$

After this rotation, the equation becomes:

$$
A'x'^2 + C'y'^2 + D'x' + E'y' + F = 0
$$

Where \( A' \) and \( C' \) are the new coefficients, and \( Bxy \) is removed.

#### **Step 2: Translate the Center (Shift)**

If the curve is not centered at the origin, translate the origin to the center \( (h, k) \) by using the following substitutions:

$$
x = x' + h
$$
$$
y = y' + k
$$

Substitute these into the equation and simplify. The result is a shifted center at \( (h, k) \).

#### **Step 3: Complete the Square**

Now, we complete the square for the quadratic terms in \( x \) and \( y \) to get the canonical form.

For example, for a circle:

$$
(x - h)^2 + (y - k)^2 = r^2
$$

For an ellipse, hyperbola, or parabola, the terms are adjusted accordingly.

### 3. **Canonical Forms After Reduction**

- **Circle**: \( (x - h)^2 + (y - k)^2 = r^2 \)
- **Ellipse**: \( \frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1 \)
- **Hyperbola**: \( \frac{(x - h)^2}{a^2} - \frac{(y - k)^2}{b^2} = 1 \)
- **Parabola**: \( y - k = a(x - h)^2 \) or \( x - h = a(y - k)^2 \)

### 4. **Conditions for Different Conics**

- **Circle**: \( A = C \) and \( B = 0 \)
- **Ellipse**: \( A \neq C \), \( B = 0 \), and \( A, C > 0 \)
- **Hyperbola**: \( B^2 - 4AC > 0 \)
- **Parabola**: \( B^2 - 4AC = 0 \)

```python
#6
import numpy as np
import matplotlib.pyplot as plt

# Hyperbola equation: x^2 - y^2 = 1

# Define x and y ranges
x = np.linspace(-3, 3, 500)
y_positive = np.sqrt(x**2 - 1)
y_negative = -np.sqrt(x**2 - 1)

# Plotting the hyperbola
plt.figure(figsize=(6, 6))
plt.plot(x, y_positive, label="Upper Branch")
plt.plot(x, y_negative, label="Lower Branch")

plt.axvline(x=0, color="red", linestyle="--", label="x = 0 (closest point)")

plt.scatter([0], [1], color="green", label="(0, 1) (Upper Branch)")
plt.scatter([0], [-1], color="blue", label="(0, -1) (Lower Branch)")

plt.text(0.1, 0, "Distance = 2", fontsize=12, color="black")

plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.title("Graph of the Hyperbola x² - y² = 1", fontsize=14)
plt.xlabel("x-axis", fontsize=12)
plt.ylabel("y-axis", fontsize=12)
plt.legend()
plt.grid(True)

plt.gca().set_aspect('equal', adjustable='box')

plt.show()

distance = 2 * 1
print(f"The distance between the branches of the hyperbola is {distance}.")
```

```python
#5
import numpy as np
import matplotlib.pyplot as plt

def hyperbola(x):
  return np.sqrt(x**2 - 1)

def ellipse(x):
  return np.sqrt((6 - x**2) / 4)

# Generate x values for plotting
x = np.linspace(-3, 3, 500)

y_hyperbola_pos = hyperbola(x)
y_hyperbola_neg = -hyperbola(x)  # Negative branch of hyperbola
y_ellipse_pos = ellipse(x)
y_ellipse_neg = -ellipse(x)

plt.figure(figsize=(6, 6))
plt.plot(x, y_hyperbola_pos, label="x² - y² = 1", color="blue")
plt.plot(x, y_hyperbola_neg, color="blue")  # Plot the negative branch
plt.plot(x, y_ellipse_pos, label="x² + 4y² = 6", color="red")
plt.plot(x, y_ellipse_neg, color="red")  # Plot the lower half

intersection_points = [(np.sqrt(2), 1), (-np.sqrt(2), 1),
                       (np.sqrt(2), -1), (-np.sqrt(2), -1)]
for point in intersection_points:
  plt.scatter(point[0], point[1], color="green", marker="o", s=50)

plt.title("Intersection of Hyperbola and Ellipse")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show()

print("Intersection points:")
for point in intersection_points:
  print(f"({point[0]:.4f}, {point[1]:.4f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
def circle(x):
  return np.sqrt(1 - x**2)
def line(x, m):
  return m*x - 5
m = np.sqrt(24)
x = np.linspace(-1, 1, 100)
y_circle = circle(x)
y_line = line(x, m)
plt.figure(figsize=(6, 6))
plt.plot(x, y_circle, label="x^2 + y^2 = 1")
plt.plot(x, -y_circle, label="x^2 + y^2 = 1")
plt.plot(x, y_line, label=f"y = {m:.2f}x - 5")
plt.title("circle and tangent line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.show()
print(f"The slope (m > 0) of the tangent line is: {m:.2f}")
```

```python
#3
import matplotlib.pyplot as plt
import numpy as np

# Ellipse equation: (x-2)^2 / 4 + (y-2)^2 / 1 = 1
# The center is (2, 2), semi-major axis a = 2, semi-minor axis b = 1.

# Define the center and axes
center_x, center_y = 2, 2
a = 2  # Semi-major axis
b = 1  # Semi-minor axis

# Generate points for the ellipse
theta = np.linspace(0, 2 * np.pi, 500)  # Parametric angle
x = center_x + a * np.cos(theta)       # Parametric equation for x
y = center_y + b * np.sin(theta)       # Parametric equation for y

# Plot the ellipse
plt.figure(figsize=(6, 6))
plt.plot(x, y, label=r"$\frac{(x-2)^2}{4} + \frac{(y-2)^2}{1} = 1$", color="blue")
plt.scatter(center_x, center_y, color="red", label="Center (2, 2)", zorder=5)

# Formatting the plot
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.gca().set_aspect("equal", adjustable="box")
plt.title("Graph of the Ellipse", fontsize=14)
plt.xlabel("x-axis", fontsize=12)
plt.ylabel("y-axis", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```
