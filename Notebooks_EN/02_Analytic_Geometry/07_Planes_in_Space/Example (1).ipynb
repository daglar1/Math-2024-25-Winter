# Planes and Lines in Space

## 1. **Planes in Space**

A **plane** in 3D space is a flat, two-dimensional surface that extends infinitely. It can be defined by a point and a normal vector perpendicular to the surface.

### **Equation of a Plane**

The general equation of a plane is given by:
$$ Ax + By + Cz = D $$

Where:
- $(A, B, C)$ is the normal vector to the plane,
- $(x, y, z)$ are the coordinates of any point on the plane,
- $D$ is a constant that determines the position of the plane.

### **Vector Equation of a Plane**

A plane can also be represented using vectors. The vector equation of a plane is:
$$ \mathbf{r} = \mathbf{r_0} + \lambda \mathbf{v_1} + \mu \mathbf{v_2} $$

Where:
- $\mathbf{r}$ is the position vector of any point on the plane,
- $\mathbf{r_0}$ is the position vector of a known point on the plane,
- $\mathbf{v_1}$ and $\mathbf{v_2}$ are two non-parallel vectors lying on the plane,
- $\lambda$ and $\mu$ are scalar parameters.

### **Plane Equation Passing Through Three Points**

Given three points $\mathbf{P_1}(x_1, y_1, z_1)$, $\mathbf{P_2}(x_2, y_2, z_2)$, and $\mathbf{P_3}(x_3, y_3, z_3)$ on the plane, we first calculate the vectors:
$$ \mathbf{v_1} = \mathbf{P_2} - \mathbf{P_1} $$
$$ \mathbf{v_2} = \mathbf{P_3} - \mathbf{P_1} $$

Then the normal vector $\mathbf{n}$ is given by the cross product:
$$ \mathbf{n} = \mathbf{v_1} \times \mathbf{v_2} $$

The equation of the plane is then:
$$ \mathbf{n} \cdot (\mathbf{r} - \mathbf{P_1}) = 0 $$

### **Plane Equation in Intercept Form**

The intercept form of the plane equation is:
$$ \frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1 $$

Where $a$, $b$, and $c$ are the intercepts on the x, y, and z axes, respectively.

### **Special Cases of Plane Positioning**

- **Parallel Planes**: Two planes are parallel if their normal vectors are proportional:
  $$ \mathbf{n_1} \parallel \mathbf{n_2} $$

- **Perpendicular Planes**: Two planes are perpendicular if their normal vectors are orthogonal:
  $$ \mathbf{n_1} \cdot \mathbf{n_2} = 0 $$

## 2. **Angle Between Two Planes**

The angle $\theta$ between two planes is determined by the angle between their normal vectors:
$$ \cos(\theta) = \frac{\mathbf{n_1} \cdot \mathbf{n_2}}{|\mathbf{n_1}| |\mathbf{n_2}|} $$

Where:
- $\mathbf{n_1}$ and $\mathbf{n_2}$ are the normal vectors of the two planes.

## 3. **Conditions for Parallelism and Perpendicularity of Two Planes**

- **Parallelism**: Two planes are parallel if their normal vectors are proportional:
  $$ \mathbf{n_1} \parallel \mathbf{n_2} $$

- **Perpendicularity**: Two planes are perpendicular if their normal vectors are orthogonal:
  $$ \mathbf{n_1} \cdot \mathbf{n_2} = 0 $$

## 4. **Line in Space**

A line in space can be represented by a parametric equation:
$$ \mathbf{r}(t) = \mathbf{r_0} + t\mathbf{v} $$

Where:
- $\mathbf{r_0}$ is a point on the line,
- $\mathbf{v}$ is the direction vector of the line,
- $t$ is a scalar parameter.

## 5. **Types of Line Equations in Space**

- **Parametric Form**:
  $$ \mathbf{r}(t) = \mathbf{r_0} + t\mathbf{v} $$

- **Symmetric Form**:
  $$ \frac{x - x_1}{a} = \frac{y - y_1}{b} = \frac{z - z_1}{c} $$

Where $(x_1, y_1, z_1)$ is a point on the line and $(a, b, c)$ is the direction vector.

## 6. **Relative Positions of a Line and a Plane**

The relative positions of a line and a plane can be:
1. **Intersecting**: The line intersects the plane at a single point.
2. **Parallel**: The line is parallel to the plane and does not intersect it.
3. **Coincident**: The line lies entirely on the plane.

## 7. **Relative Positions of Two Lines**

Two lines in space can be:
1. **Intersecting**: The lines intersect at a single point.
2. **Parallel**: The lines do not intersect and have the same direction.
3. **Skew**: The lines do not intersect and are not parallel (they lie in different planes).

## 8. **Angle Between a Line and a Plane**

The angle $\theta$ between a line and a plane is given by the complement of the angle between the line and the normal to the plane:
$$ \sin(\theta) = \frac{|\mathbf{v} \cdot \mathbf{n}|}{|\mathbf{v}| |\mathbf{n}|} $$

Where:
- $\mathbf{v}$ is the direction vector of the line,
- $\mathbf{n}$ is the normal vector of the plane.

```python
#9
import numpy as np

# Define the normal vectors of the planes
n1 = np.array([1, 1, 1])  # Normal of plane x + y + z = 1
n2 = np.array([1, 0, 0])  # Normal of plane x = 0 (yz-plane)

# Compute the dot product of the normal vectors
dot_product = np.dot(n1, n2)

# Compute the magnitudes of the normal vectors
magnitude_n1 = np.linalg.norm(n1)
magnitude_n2 = np.linalg.norm(n2)

# Calculate the cosine of the angle
cos_theta = abs(dot_product) / (magnitude_n1 * magnitude_n2)

# Calculate the angle in radians and convert to degrees
angle_radians = np.arccos(cos_theta)
angle_degrees = np.degrees(angle_radians)

print(f"The angle between the planes is {angle_degrees:.2f} degrees.")
```

```python
#8
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the plane equation: 6x + 4y + 3z = 12
def plane(x, y):
    return (12 - 6*x - 4*y) / 3

# Define the points A, B, C
A = np.array([2, 0, 0])
B = np.array([0, 3, 0])
C = np.array([0, 0, 4])

# Generate a grid for plotting the plane
x = np.linspace(-1, 3, 100)
y = np.linspace(-1, 4, 100)
X, Y = np.meshgrid(x, y)
Z = plane(X, Y)

# Plot the plane
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.5, color='blue')

# Plot points A, B, C
ax.scatter(*A, color='red', label='A(2, 0, 0)', s=100)
ax.scatter(*B, color='green', label='B(0, 3, 0)', s=100)
ax.scatter(*C, color='orange', label='C(0, 0, 4)', s=100)

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Plane through A(2, 0, 0), B(0, 3, 0), and C(0, 0, 4)')

# Add legend
ax.legend()

# Show plot
plt.show()
```

```python
#7
import numpy as np

# Define the plane coefficients: 2x + 3y + 4z - 5 = 0
a, b, c, d = 2, 3, 4, -5

# Define the point A(1, 2, 3)
point = np.array([1, 2, 3])

# Calculate the distance
numerator = abs(a * point[0] + b * point[1] + c * point[2] + d)
denominator = np.sqrt(a**2 + b**2 + c**2)
distance = numerator / denominator

print(f"The distance from point A to the plane is: {distance}")
```

```python
#6
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the plane equation: 2x + 3y + 4z = 5
def original_plane(x, y):
    return (5 - 2*x - 3*y) / 4

# Parallel plane example: 2x + 3y + 4z = 10
def parallel_plane(x, y):
    return (10 - 2*x - 3*y) / 4

# Perpendicular plane example: x + y + z = 1
def perpendicular_plane(x, y):
    return 1 - x - y

# Generate meshgrid for planes
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z_original = original_plane(X, Y)
Z_parallel = parallel_plane(X, Y)
Z_perpendicular = perpendicular_plane(X, Y)

# Plot the planes
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot original plane
ax.plot_surface(X, Y, Z_original, alpha=0.5, color='blue', label="Original Plane")

# Plot parallel plane
ax.plot_surface(X, Y, Z_parallel, alpha=0.5, color='green', label="Parallel Plane")

# Plot perpendicular plane
ax.plot_surface(X, Y, Z_perpendicular, alpha=0.5, color='red', label="Perpendicular Plane")

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Planes: Original, Parallel, and Perpendicular')

# Show the plot
plt.show()
```

```python
#5
import numpy as np

# Define the two vectors
v1 = np.array([1, 0, 1])
v2 = np.array([0, 1, -1])

# Compute the normal vector (cross product)
normal_vector = np.cross(v1, v2)

# Define the point A(1, 2, 3)
point = np.array([1, 2, 3])

# Find the equation of the plane
# General form: n1*(x - x0) + n2*(y - y0) + n3*(z - z0) = 0
# Expand: n1*x + n2*y + n3*z - (n1*x0 + n2*y0 + n3*z0) = 0
d = -np.dot(normal_vector, point)
plane_equation = f"{normal_vector[0]}x + {normal_vector[1]}y + {normal_vector[2]}z + {d} = 0"

print("Equation of the plane:")
print(plane_equation)
```

```python
#4
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the normal vectors of the planes
n1 = np.array([2, 3, 4])  # Normal of plane 1
n2 = np.array([3, 4, 2])  # Normal of plane 2

# Compute the direction vector of the line (cross product of normals)
direction_vector = np.cross(n1, n2)

# Solve the system of equations to find a point on the line
# Plane equations: 2x + 3y + 4z = 5 and 3x + 4y + 2z = 6
# We'll set z = 0 (arbitrary choice) and solve for x and y
from sympy import symbols, Eq, solve

x, y, z = symbols('x y z')
plane1 = Eq(2*x + 3*y + 4*z, 5)
plane2 = Eq(3*x + 4*y + 2*z, 6)

solution = solve([plane1, plane2, Eq(z, 0)], (x, y, z))
point_on_line = np.array([float(solution[x]), float(solution[y]), float(solution[z])])

# Generate points along the line
t = np.linspace(-10, 10, 100)
line_points = point_on_line + t[:, None] * direction_vector

# Plot the planes and the line of intersection
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Define a grid for the planes
xx, yy = np.meshgrid(np.linspace(-10, 10, 20), np.linspace(-10, 10, 20))
plane1_z = (5 - 2*xx - 3*yy) / 4
plane2_z = (6 - 3*xx - 4*yy) / 2

# Plot the planes
ax.plot_surface(xx, yy, plane1_z, alpha=0.5, color='blue', label='Plane 1')
ax.plot_surface(xx, yy, plane2_z, alpha=0.5, color='green', label='Plane 2')

# Plot the line of intersection
ax.plot(line_points[:, 0], line_points[:, 1], line_points[:, 2], color='red', label='Line of Intersection')

# Add labels and legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Line of Intersection of Two Planes')
ax.legend()

# Show the plot
plt.show()
```
