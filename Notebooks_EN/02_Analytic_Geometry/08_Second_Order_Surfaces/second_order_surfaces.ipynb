## 1. General Equations of 3D Surfaces

### 1.1 Ellipsoid
The general equation of an ellipsoid centered at the origin is:
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2} = 1
$$
where:
- $a$, $b$, and $c$ are the semi-axes of the ellipsoid along the $x$, $y$, and $z$ axes, respectively.

#### Special Case: Sphere
When $a = b = c = r$, the ellipsoid becomes a sphere:
$$
x^2 + y^2 + z^2 = r^2
$$
where $r$ is the radius of the sphere.

---

### 1.2 Hyperboloid
#### One-Sheeted Hyperboloid
The general equation of a hyperboloid of one sheet is:
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1
$$
where:
- $a$, $b$, and $c$ are constants determining the shape and orientation of the hyperboloid.

#### Two-Sheeted Hyperboloid
The general equation of a hyperboloid of two sheets is:
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = -1
$$

For hyperboloids, the sign of the constant on the right-hand side determines whether the surface is one-sheeted or two-sheeted.

---

### 1.3 Sphere
A sphere is a special case of the ellipsoid where all semi-axes are equal:
$$
x^2 + y^2 + z^2 = r^2
$$
where $r$ is the radius of the sphere.

For a sphere with its center shifted to $(h, k, l)$:
$$
(x - h)^2 + (y - k)^2 + (z - l)^2 = r^2
$$

```python
#4
from sympy import symbols, diff, Eq

# Define the variables
x, y, z = symbols('x y z')

# Define the paraboloid equation
paraboloid = (x - 1)**2 + y**2 + 1

# Define the point of tangency
P = (1, 0, 1)

# Calculate the partial derivatives
partial_x = diff(paraboloid, x)
partial_y = diff(paraboloid, y)

# Evaluate the partial derivatives at point P
partial_x_at_P = partial_x.subs({x: P[0], y: P[1]})
partial_y_at_P = partial_y.subs({x: P[0], y: P[1]})

# Tangent plane equation
# z - z0 = fx(x - x0) + fy(y - y0)
tangent_plane = Eq(z - P[2], partial_x_at_P * (x - P[0]) + partial_y_at_P * (y - P[1]))

tangent_plane
```

```python
#3
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parameters of the curve
theta = np.linspace(0, 2 * np.pi, 100)

# The curve lies in the plane x = 0.5
x = 0.5 * np.ones_like(theta)

# The circle in the y-z plane at x = 0.5
radius = np.sqrt(0.75)  # Derived from x^2 + y^2 + z^2 = 1
y = radius * np.cos(theta)
z = radius * np.sin(theta)

# Plot the curve
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the spheres for context
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
X = np.outer(np.cos(u), np.sin(v))
Y = np.outer(np.sin(u), np.sin(v))
Z = np.outer(np.ones(np.size(u)), np.cos(v))

# Sphere 1
ax.plot_surface(X, Y, Z, color='blue', alpha=0.3, label="Sphere 1")

X2 = X + 1
ax.plot_surface(X2, Y, Z, color='green', alpha=0.3, label="Sphere 2")

# Plot the curve of intersection
ax.plot(x, y, z, color='red', linewidth=2, label="Intersection Curve")

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Intersection of Two Spheres')
ax.legend()
plt.show()
```

```python
#2
# Python code to determine if two spheres have common points

# Sphere 1: x^2 + y^2 + z^2 = 1
# Sphere 2: x^2 + y^2 + z^2 = 2

# The spheres are concentric and have different radii.
# Let's determine if they have common points geometrically.

# Define sphere radii
radius1 = 1  # Sphere 1 radius
radius2 = 2  # Sphere 2 radius

# Define sphere centers
center1 = (0, 0, 0)  # Center of Sphere 1
center2 = (0, 0, 0)  # Center of Sphere 2

# Calculate the distance between centers
distance_between_centers = ((center1[0] - center2[0])**2 +
                            (center1[1] - center2[1])**2 +
                            (center1[2] - center2[2])**2)**0.5

# Determine if spheres intersect
if distance_between_centers < abs(radius1 - radius2):
    print("The spheres do not intersect; one sphere is contained within the other.")
elif distance_between_centers > (radius1 + radius2):
    print("The spheres do not intersect; they are separate.")
elif distance_between_centers == 0 and radius1 != radius2:
    print("The spheres are concentric and do not intersect.")
else:
    print("The spheres intersect or are tangent.")

```
